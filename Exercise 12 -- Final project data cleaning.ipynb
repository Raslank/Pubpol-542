```python
badENVsymbols=[] # list for bad symbols

NumericColNames=environment.iloc[:,1:].columns # save names of columns with numeric data

for columnName in NumericColNames:# visit every column name
    for cell in environment[columnName]:# visit every cell for that column
        try:
            float(cell) # try this
        except: # if not possible:            
            if cell not in badENVsymbols:# if cell is not in the list                
                badENVsymbols.append(cell)# add it to the list

# you get:
badENVsymbols
#shows all of the numeric columns
#that have non-numeric values or symbols
```

```python
environment.replace(to_replace=badENVsymbols,value=np.nan,inplace=True)
```

```python
# getting the column that we want to work with
num = 1
ey = 1992
col = e.iloc[:,num]

# for each row in col (x) 
for x in range(len(col)):
    
    
    # cell value because we are going down column 1 looking at each row (x)
    cell_value = e.iloc[x,num]
    # convert each cell value to a float because some are not showing as numbers
    year = float(cell_value)
    
    if year <= ey:
        e.iloc[x,num] = "Original Adopter"
    
    
    # create conditions 
    if year > ey and year <= ey+5:
        
     # if it meets above criteria it will be changed to what's in the quotes changing value of variable not actual location so you need to have the e.loc statement
        e.iloc[x,num] = "1-5 Year Adopter"
        
    if year > ey+5 and year <= ey+10 :
        e.iloc[x,num] = "6-10 year adopter"
        
    if year > ey+10:
        e.iloc[x,num] = "11+ year adopter"
        
    if year != year:
        e.iloc[x,num] = "Non-adopter"
```
